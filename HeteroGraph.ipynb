```python
def load_labels(path):
    with open('labels_list_dict.pkl', 'rb') as f:
        labels_list = pickle.load(f)
    
    return labels_list


def explode_labels(labels_list):
    exploded_labels = pd.Series(labels_list).explode()
    exploded_labels_df = exploded_labels.to_frame()
    exploded_labels_df.columns = ['label']

    return exploded_labels_df


def get_unique_labels(y):

    unique_side_effects = y.drop_duplicates()
    print("Number of Unique Side Effects: ", unique_side_effects.shape)



    side_effect_mapping = {}
    for idx, row in enumerate(unique_side_effects):
        side_effect_mapping[row] = idx

    side_effect_mapping

    side_effects = pd.DataFrame(side_effect_mapping.keys(), columns=['side_effect_name'])
    side_effects.index = side_effect_mapping.values()

    return side_effects, side_effect_mapping



labels_list = load_labels('labels_list_dict.pkl')

exploded_labels_df = explode_labels(labels_list)


ungrouped_df = pd.concat([pdrugs_drugid.reset_index(drop=True), exploded_labels_df], axis=1).reset_index(drop=True)
y_ungrouped_df = ungrouped_df.iloc[:, -1]


side_effects, side_effect_mapping = get_unique_labels(y_ungrouped_df)

print(y_ungrouped_df)
se_unique_mapping = y_ungrouped_df.map(side_effect_mapping)
```

```python
def _split(
    store,
    index,
    is_undirected: bool,
    rev_edge_type,
):

    edge_attrs = {key for key in store.keys() if store.is_edge_attr(key)}
    for key, value in store.items():
        if key == 'edge_index':
            continue

        if key in edge_attrs:
            value = value[index]
            if is_undirected:
                value = torch.cat([value, value], dim=0)
            store[key] = value

    edge_index = store.edge_index[:, index]
    if is_undirected:
        edge_index = torch.cat([edge_index, edge_index.flip([0])], dim=-1)
    store.edge_index = edge_index

    if rev_edge_type is not None:
        rev_store = store._parent()[rev_edge_type]
        for key in rev_store.keys():
            if key not in store:
                del rev_store[key]  # We delete all outdated attributes.
            elif key == 'edge_index':
                rev_store.edge_index = store.edge_index.flip([0])
            else:
                rev_store[key] = store[key]

    return store
```

```python
def _create_label(
    store,
    index,
    neg_edge_index,
    out,
):
    key = 'edge_label'
    edge_index = store.edge_index[:, index]

    if hasattr(store, key):
        edge_label = store[key]
        edge_label = edge_label[index]
        # Increment labels by one. Note that there is no need to increment
        # in case no negative edges are added.
        if neg_edge_index.numel() > 0:
            assert edge_label.dtype == torch.long
            assert edge_label.size(0) == edge_index.size(1)
            edge_label.add_(1)
        if hasattr(out, key):
            delattr(out, key)
    else:
        edge_label = torch.ones(index.numel(), device=index.device)

    if neg_edge_index.numel() > 0:
        neg_edge_label = edge_label.new_zeros((neg_edge_index.size(1), ) +
                                                edge_label.size()[1:])

    if neg_edge_index.numel() > 0:
        edge_label = torch.cat([edge_label, neg_edge_label], dim=0)
        edge_index = torch.cat([edge_index, neg_edge_index], dim=-1)
    out[key] = edge_label
    out[f'{key}_index'] = edge_index

    return out
```

```python
from torch_geometric.nn import SAGEConv, to_hetero, GCNConv, GraphConv, GATConv
import torch.nn.functional as F
import torch
from torch import Tensor

class GNN(torch.nn.Module):
    def __init__(self, hidden_channels, dropout_rate = 0.2):
        super().__init__()

        self.conv1 = GraphConv(hidden_channels, hidden_channels)
        self.conv2 = GraphConv(hidden_channels, hidden_channels)
        self.conv3 = GraphConv(hidden_channels, hidden_channels)
        self.dropout = torch.nn.Dropout(dropout_rate)

    def forward(self, x: Tensor, edge_index: Tensor) -> Tensor:
        x = self.conv1(x, edge_index)
        x = F.leaky_relu(x)
        x = self.dropout(x)
        x = self.conv2(x, edge_index)
        x = F.leaky_relu(x)
        x = self.dropout(x)
        x = self.conv3(x, edge_index)
        return x

# Our final classifier applies the dot-product between source and destination
# node embeddings to derive edge-level predictions:
class Classifier(torch.nn.Module):
    def forward(self, x_pdrugs: Tensor, x_seffect: Tensor, edge_label_index: Tensor) -> Tensor:
        # Convert node embeddings to edge-level representations:
        edge_feat_pdrugs = x_pdrugs[edge_label_index[0]]
        edge_feat_seffect = x_seffect[edge_label_index[1]]

        # Apply dot-product to get a prediction per supervision edge:
        return (edge_feat_pdrugs * edge_feat_seffect).sum(dim=-1)


class Model(torch.nn.Module):
    def __init__(self, hidden_channels, pdrugs_size):
        super().__init__()
        # Since the dataset does not come with rich features, we also learn two
        # embedding matrices for pdrugss and seffects:
        self.seffect_lin = torch.nn.Linear(768, hidden_channels)
        self.pdrugs_lin = torch.nn.Linear(pdrugs_size, hidden_channels)
        self.pdrugs_emb = torch.nn.Embedding(data["pdrugs"].num_nodes, hidden_channels)
        self.seffect_emb = torch.nn.Embedding(data["seffect"].num_nodes, hidden_channels)

        # Instantiate homogeneous GNN:
        self.gnn = GNN(hidden_channels)

        # Convert GNN model into a heterogeneous variant:
        self.gnn = to_hetero(self.gnn, metadata=data.metadata())

        self.classifier = Classifier()

    def forward(self, data: HeteroData, is_neg_sampling = True) -> Tensor:
        x_dict = {
          "pdrugs": self.pdrugs_lin(data["pdrugs"].x.float()) + self.pdrugs_emb(data["pdrugs"].node_id),
          "seffect": self.seffect_lin(data["seffect"].x.float()) + self.seffect_emb(data["seffect"].node_id),
        } 
        

        x_dict = self.gnn(x_dict, data.edge_index_dict)

        if is_neg_sampling == True:
            neg_edge_index = negative_sampling(
                edge_index=data["pdrugs", "associated", "seffect"].edge_index, num_nodes=(data['pdrugs'].num_nodes, data['seffect'].num_nodes),
                num_neg_samples=data["pdrugs", "associated", "seffect"].edge_label_index.size(1), method='dense')
        
            edge_label_index = torch.cat(
                [data["pdrugs", "associated", "seffect"].edge_label_index, neg_edge_index],
                dim=-1,
            )

            edge_label = torch.cat([
                data["pdrugs", "associated", "seffect"].edge_label,
                data["pdrugs", "associated", "seffect"].edge_label.new_zeros(neg_edge_index.size(1))
            ], dim=0)
            
        elif is_neg_sampling == False:
            edge_label_index = data["pdrugs", "associated", "seffect"].edge_label_index
            edge_label = data["pdrugs", "associated", "seffect"].edge_label

        # `x_dict` holds feature matrices of all node types
        # `edge_index_dict` holds all edge indices of all edge types
        pred = self.classifier(
            x_dict["pdrugs"],
            x_dict["seffect"],
            edge_label_index 
        )

        return pred, edge_label


def log_metrics(metric_values, metrics_names, model_name, columns=[], values=[], filepath = 'evaluation_results2.csv' ):

    file_exists = os.path.isfile(filepath)

    # Create or open the file and write the results
    with open(filepath, mode='a', newline='') as file:
        writer = csv.writer(file)
        
        if not file_exists:
            writer.writerow(['Model', 'Metric', 'Value'] + columns)
        
        # Write the test metrics
        for metric_name, metric_value in zip(metrics_names, metric_values):
            writer.writerow([model_name, metric_name, metric_value] + values)
```
